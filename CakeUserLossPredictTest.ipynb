```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
% matplotlib inline
```

```python
data = pd.read_excel('/Users/pein/Desktop/12月份蛋糕消费数据集(1).xlsx')
```

```python
data.head()
```

```python
# 将数据切分成两段
target = data['Loss_or_not']
df = data.iloc[:,:-1]
```

```python
# 基本的数据处理，转化格式。去除uid列
df['provinceId'] = df.provinceId.astype('category')
df['jude_value'] = df.jude_value.astype(float)
df = df.drop(['uid'],1)
```

数据的预处理

```python
#处理出生年龄
df['Age'] = 2017-df.birthYear
mask_age_lower = df['Age']<10
mask_age_high = df['Age']>90
df['Age_right'] =1
df['Age_right'][mask_age_lower] = 0
df['Age_right'][mask_age_high] = 0
df['Age'][mask_age_lower] = 0
df['Age'][mask_age_high] = 0
```

```python
# 处理性别数据
mask_gender_error = df['gender']==-1
df['gender_right'] = 1
df['gender_right'][mask_gender_error] =0
```

```python
df.describe()
```

```python
# 观察数据后，删除birthYear列
#df = df.drop(['birthYear','jude_value','register_channelId',
              #'cunpon_use','jifen_use','cancel_order',
             #'GoodsView','ActiveDays','Age','gender'],1)
df = df.drop(['birthYear'],1)
```

```python
# 数据的标准化
df.iloc[:,1:] = df.iloc[:,1:].apply(lambda x: (x-np.mean(x))/np.std(x))
df.head()
```

```python
#  将数据划分为训练集和测试集,按照7:3的比例来划分训练集和测试集
df_train = df.iloc[:15000,:]
df_test = df.iloc[15000:,:]
target_train = target[:15000]
target_test = target[15000:]

df_train.head()
```

```python
#采用决策树模型
```

```python
# 采用LR模型和SVM模型
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import svm
# LR模型的效果
LR = LogisticRegression()
LR.fit(df_train,target_train)
score = LR.score(df_train,target_train)
score
```

```python
predict = LR.predict(df_test)
accuracy_score(target_test,predict)
```

```python
# SVM模型的效果
SVM = svm.SVC(kernel='sigmoid',C=0.1)
SVM.fit(df_train, target_train)
score1 = SVM.score(df_train, target_train)
score1
```

```python
predict1 = SVM.predict(df_test)
accuracy_score(target_test, predict1)
```

```python
# 随机森林算法效果
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(n_estimators=20)
RFC.fit(df_train,target_train)
score2 = RFC.score(df_train,target_train)
score2
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
predict2 = RFC.predict(df_test)
accuracy_score(target_test, predict2)
```

```python
# 测试召回率，综合评价模型的准确预测效果
recall_score(target_test, predict2,average='binary')
```

```python
# 测试精确度，综合评价模型的准确性
precision_score(target_test, predict2,average='binary')
```
