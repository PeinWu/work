```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
% matplotlib inline
```

```python
data = pd.read_excel('/Users/pein/Desktop/data(1).xlsx')
```

```python
data.head()
```

```python
data['Age_right'] = 0
correct = ((data.Age>10) & (data.Age<90))
data.Age_right[correct] = 1
```

```python
data.head()
```

```python
# Use mask in the replace with the outliers
outlier = (data.Age<10)
data['Age'][outlier] = 25
outlier1 = (data.Age>90)
data['Age'][outlier1] = 25
```

```python
data.describe()
```

```python
data.drop('createOn',1)
```

```python
data = data[['uid', 'provinceId', 'gender', 'gender_right', 'Age', 'createOn_Timerange',
            'orderNum', 'LastConsuMe_TimeRange','totalFee', 'SingtotalFee',
            'ActiveDays','ViewGoodsNum','Age_right','ConsumeOrNot']]
```

```python
data.info()
```

```python
data.drop('uid',1)
```

```python
data_train = data.iloc[:,1:-1]
data_test = data.iloc[:,-1]
```

```python
data_train.info()
```

```python
data_train['provinceId'] = data_train['provinceId'].astype('category')
```

```python
data_train['ViewGoodsNum'] = data_train['ViewGoodsNum'].fillna(1)
data_train.info()
```

```python
data_train.iloc[:,1:] = data_train.iloc[:,1:].apply(lambda x: (x-np.mean(x))/np.std(x))
data_train.head()
```

```python
# the LR model for the consumerLoss
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

print(data_train.head())
clf = LogisticRegression()
clf.fit(data_train, data_test)

result = clf.score(data_train,data_test)
result
```

```python
data_train_train = data_train.iloc[:15000,:]
data_train_test = data_train.iloc[15000:,:]
data_test_train = data_test.iloc[:15000]
data_test_test = data_test.iloc[15000:]
```

```python
clf1 = LogisticRegression()
clf1.fit(data_train_train, data_test_train)

predict = clf1.predict(data_train_test)
accuracy_score(data_test_test,predict)
```

```python
# the SVM model for consumerLoss
from sklearn import svm
clf2 = svm.SVC()
clf2.fit(data_train_train, data_test_train)
result1 = clf2.score(data_train_train, data_test_train)
result1
```

```python
predict2 = clf2.predict(data_train_test)
accuracy_score(data_test_test,predict2)
```
